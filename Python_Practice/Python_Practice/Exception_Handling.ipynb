```python
#Try , Except and Finally

#Catching Exception in Python

import sys   #sys module helps get the type of exception

mylist=['abc',0,4]

for value in mylist:
    try:
        print('Element from the list is :',value)
        inverse=1/value
    except:
        print("Error observed: ", sys.exc_info()[0],'occured' )
        print("------------------------------------------------------")
        
print("The inverse of ",value, "is :", inverse)
```

```python
#Catching Specific Exception in Python

import sys   #sys module helps get the type of exception

mylist=['abc',0,4]

for value in mylist:
    try:
        print("--------------------------------------------------")
        print('Element from the list is :',value)
        inverse=1/value
        
    except(TypeError):
        print("Type Error occured you can not get inverse of a string")
    except(ZeroDivisionError):
        print(" Any value can not be divided with zero")
    except:
        print("Generic Error")
        
        
        
print("The inverse of ",value, "is :", inverse)

```
